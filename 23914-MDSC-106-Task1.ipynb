```python
def throwPie(df, columns):
    all_labels = []
    all_sizes = []
    all_titles = []
    for column in columns:
        series = df[column].value_counts()
        sizes = []
        labels = []
        titles = []
        titles.append(column)
        for x in range(len(series.index)):
            labels.append(series.index[x])
        for attr in series:
            sizes.append(attr)
        all_titles.append(titles)
        all_sizes.append(sizes)
        all_labels.append(labels)
        # ax[index].set_title(column)
        # ax[index].pie(sizes, labels=labels, autopct='%1.1f%%')
    rows = math.ceil(len(columns) / 3 )
    cols = math.ceil(len(columns) / rows)
    fig, ax = plt.subplots(rows, cols, figsize=(30, 30))
    count = 0
    for row in range(rows):
        for col in range(cols):
            if count >= len(columns):
                break
            fig.suptitle("Analysing Categorical Features")
            ax[row,col].set_title(all_titles[count])
            ax[row, col].pie(all_sizes[count], labels=all_labels[count], autopct='%1.1f%%')
            count += 1
            
    fig.savefig("piechart")
            
        
        
```

```python
columns = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn']
```

```python
throwPie(telecom_data, columns)
```

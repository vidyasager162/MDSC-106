```python
date = us_unemployment['date'].tail(25)
age_20_24 = us_unemployment['age_20_24_rate'].tail(25)
age_25_34 = us_unemployment['age_25_34_rate'].tail(25)
plt.figure(figsize=(25,10))
plt.title("US Unemployment 2021-2023 Post Covid Analysis")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate")
plt.scatter(date, age_20_24, label="Age: 20-24")
plt.scatter(date, age_25_34, label="Age: 25-34")
plt.legend()
```

From this scatter plot we can clearly see that in the United States, the unemployment rates have been improving post covid (From 2021 - 2023).

More specifically, the Unemployment Rate for people in the age group 20-24 was higher than for the people in the age group 25-34. But as time went on, the rates improved and by 2023-07-01, the rate of unemployment for people of age group 25-34 went to less than 4 and for the age group 20-24 it came down to 7.

```python
print("The Highest Rate for people of age group 20-24 was: ", age_20_24.max())
print("The Lowest Rate for people of age group 20-24 was: ", age_20_24.min())
print("The Highest Rate for people of age group 25-34 was: ", age_25_34.max())
print("The Lowest Rate for people of age group 25-34 was: ", age_25_34.min())
```

The same is represented in a Bar Plot

```python
us_unemployment = us_unemployment.tail(25)
sns.set(rc={"figure.figsize": (20,10)})
sns.barplot(data=us_unemployment, x="date", y="age_20_24_rate")
```

The Unemployment Rates from 2021 to 2023 for the age groups 20-24,25-34,25-54 and 35-44 is represented as a stacked bar plot

```python
us_unemployment1 = us_unemployment.drop(['overall_rate','age_16_17_rate','age_16_19_rate','age_18_19_rate','age_16_24_rate','age_45_54_rate','age_20plus_rate','age_25plus_rate','age_55plus_rate'],axis=1)
us_unemployment1.plot(kind="bar", stacked="True")
```

The count of people with specific unemployment rates in the age group of 20-24 is shown using the following Histogram plot.

```python
us_gender_based_path = "./archive/df_sex_unemployment_rates.csv"
us_gender_based = pd.read_csv(us_gender_based_path)
us_gender_based
```

From this given dataset we will draw a few plots comparing the rate of unemployment between men and women.

```python
us_gender_based = us_gender_based.drop(['overall_rate','men_16_17_rate','women_16_17_rate','men_16_19_rate','women_16_19_rate','men_18_19_rate','women_18_19_rate','men_16_24_rate','women_16_24_rate','men_20_24_rate','women_20_24_rate','men_25plus_rate','women_25plus_rate','men_25_34_rate','women_25_34_rate','men_25_54_rate','women_25_54_rate','men_35_44_rate','women_35_44_rate','men_45_54_rate','women_45_54_rate','men_55plus_rate','women_55plus_rate'],axis=1)
us_gender_based
```

```python
print("The Highest Unemployment Rate for Men is: ", us_gender_based['men_rate'].max())
print("The Lowest Unemployment Rate for Men is: ", us_gender_based['men_rate'].min())
print("The Highest Unemployment Rate for Women is: ", us_gender_based['women_rate'].max())
print("The Lowest Unemployment Rate for Women is: ", us_gender_based['women_rate'].min())
```

```python
us_gender_based1 = us_gender_based.tail(25)
plt.subplot(1,2,1)
sns.barplot(x="date",y="men_rate",data=us_gender_based1)
plt.subplot(1,2,2)
sns.barplot(x="date",y="women_rate",data=us_gender_based1)
```

This two comparisons shows us the change in trends. Overall if you see the highest unemployment rate was for Women. But post covid times show that the highest unemployment rate is for Men.

```python
print("The Highest Unemployment Rate for Men (2021-2023) is: ", us_gender_based1['men_rate'].max())
print("The Lowest Unemployment Rate for Men (2021-2023) is: ", us_gender_based1['men_rate'].min())
print("The Highest Unemployment Rate for Women (2021-2023) is: ", us_gender_based1['women_rate'].max())
print("The Lowest Unemployment Rate for Women (2021-2023) is: ", us_gender_based1['women_rate'].min())
```

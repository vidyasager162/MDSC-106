```python
import pandas as pd
import sklearn
```

```python
df = pd.read_excel('./End Sem//Shopping Trends - Q2/shopping_trends.xlsx')
df.head()
```

```python
df.drop(['Customer ID'], axis=1, inplace=True)
df.head()
```

```python
df['Shipping Type'].value_counts()
```

```python
df['Gender'] = [1 if x == 'Male' else 0 for x in df['Gender']]
df['Size'] = [0 if x == 'S' else 1 if x == 'M' else 2 if x == 'L' else 3 for x in df['Size']]
df['Season'] = [0 if x == 'Spring' else 1 if x == 'Fall' else 2 if x == 'Winter' else 3 for x in df['Season']]
df['Subscription Status'] = [0 if x == 'No' else 1 for x in df['Subscription Status']]
df['Discount Applied'] = [0 if x == 'No' else 1 for x in df['Discount Applied']]
df['Promo Code Used'] = [0 if x == 'No' else 1 for x in df['Promo Code Used']]
df['Shipping Type'] = [0 if x == 'Free Shipping' else 1 if x == 'Standard' else 2 if x == 'Store Pickup' else 3 if x == 'Next Day Air' else 4 if x == 'Express' else 5 for x in df['Season']]
```

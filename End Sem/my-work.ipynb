```python
genders = pd.DataFrame(df['Gender'].value_counts())
sizes = pd.DataFrame(df['Size'].value_counts())
categories = pd.DataFrame(df['Category'].value_counts())
seasons = pd.DataFrame(df['Season'].value_counts())
subscriptions = pd.DataFrame(df['Subscription Status'].value_counts())
discounts = pd.DataFrame(df['Discount Applied'].value_counts())
promos = pd.DataFrame(df['Promo Code Used'].value_counts())
payments = pd.DataFrame(df['Payment Method'].value_counts())
frequencies = pd.DataFrame(df['Frequency of Purchases'].value_counts())
```

```python
def horizontalCount(df, feature, color, sorting='descending'):
    fig = px.histogram(df, y=feature, color=color, text_auto=True).update_yaxes(categoryorder='total ' + sorting)
    fig.show()
```

```python
def verticalCount(df, feature, color, sorting='ascending'):
    fig = px.histogram(df, x=feature, color=color, text_auto=True).update_xaxes(categoryorder='total ' + sorting)
    fig.show()
```

```python
def throwHistogram(df, feature, color=None, text_auto=False):
    fig = px.histogram(df, x=feature, color=color, text_auto=text_auto)
    fig.show()
```

```python
def hundredPerStack(df, feature, color=None, barnorm='percent', text_auto=True):
    fig = px.histogram(df, x=feature, color=color, barnorm=barnorm, text_auto=text_auto)
    fig.show()
```

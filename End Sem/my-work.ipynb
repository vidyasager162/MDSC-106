```python
for col in df:
    if df[col].dtype in ['int64', 'float64']:
        continue
    else:
        print('-' * 20)
        print(df[col].value_counts())
```

```python
df.head()
```

```python
df['Gender'] = [0 if gender == 'Male' else 1 for gender in df['Gender']]
df['Subscription Status'] = [0 if status == 'No' else 1 for status in df['Subscription Status']]
df['Discount Applied'] = [0 if x == 'No' else 1 for x in df['Discount Applied']]
df['Promo Code Used'] = [0 if x == 'No' else 1 for x in df['Promo Code Used']]
```
